```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


World = input('Input World file: ')

Him = input('Input Wally file of matching numeric value: ')

# Read in the Wheres Wally page and convert it form BGR to Gray
#Read in jpg value frm World1,jpg to World3.jpg
WheresWally = cv2.imread(World)
WheresWallyGray = cv2.cvtColor(WheresWally, cv2.COLOR_BGR2GRAY)

# Read in an image of Wally himself from that picture
#Read in values from Wally1.jpg to Wally3.jpg
WallyHimself = cv2.imread(Him,0)
```

```python
# Fill w and h with the width and height of WallyHimself and create a template
w, h = WallyHimself.shape[::-1]
res = cv2.matchTemplate(WheresWallyGray,WallyHimself,cv2.TM_CCOEFF_NORMED)
threshold = 0.6
```

```python
# Finding the edges of the template
loc = np.where( res >= threshold)
for pt in zip(*loc[::-1]):
    # Placing a box around the area
    cv2.rectangle(WheresWally, pt, (pt[0] + w, pt[1] + h), (0,255,0), 2)
cv2.imwrite('found_wally.png',WheresWally)
```

```python
result = cv2.imread('found_wally.png',0)

plt.imshow(result, cmap = 'gray')

print ('File can be found in the directory used to run program under the name "found_wally.png"')
```
